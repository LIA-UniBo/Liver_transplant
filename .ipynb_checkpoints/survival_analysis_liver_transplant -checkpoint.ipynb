## Preprocessing 

```python
df = pd.read_excel('data.xlsx', engine='openpyxl')
df.head(10)
```

Measuring ratio of missing values per column

```python
missing_percent = {}
for column in df.columns:
    missing_percent[column] = df[column].isna().sum() / len(df)
for column in sorted(missing_percent, key=missing_percent.get, reverse=True):
    print('{}: {:.3f}'.format(column, missing_percent[column]))
```

We remove every attribute with 20% of missing values

```python
drop_list = [k for k, v in missing_percent.items() if v >= 0.198 ]
print(drop_list)
df = df.drop(drop_list, axis=1)
```

Encoding categorical attributes (One-Hot-Encoding)

```python
categories = ['pers_sesso', 'ric_HCV', 'HCC', 'trombosi_portale',
              'ric_HBsAg', 'UNOS', 'ric_HIV', 'chirurgia_addom', 'iscriz_tx_comb']
```

```python
df = pd.get_dummies(df, columns=categories, drop_first=True)
df
```
